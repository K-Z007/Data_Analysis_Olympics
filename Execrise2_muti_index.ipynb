# Exercise 2: Get a Table of Country Ranked by Total Olympic Medals Won over Population Size in 2014 Summer

```python
import pandas as pd
import numpy as np
```

```python
summer_games = pd.read_csv("summer_games.csv")
winter_games = pd.read_csv("winter_games.csv")
countries = pd.read_csv("countries.csv")
athletes = pd.read_csv("athletes.csv")
country_stats = pd.read_csv("country_stats.csv")
```

```python
display(summer_games.head(2))
display(winter_games.head(2))
display(countries.head(2))
display(athletes.head(2))
display(country_stats.head(2))
```

```python
country_stats['pop_in_millions'] = country_stats['population'] / 1000000
```

```python
countries['country_code'] = countries['country'].str.strip().str.upper().str.replace('.','', regex=False).str.slice(0,3)
```

```python
country_pop_gold = summer_games.merge(countries, left_on='country_id', right_on='id')\
            .merge(country_stats, left_on=['country_id', 'year'], right_on=['country_id', 'year'])\
            .groupby(['country_code', 'pop_in_millions'])[['gold','silver','bronze']].sum()
```

```python
country_pop_gold
```

```python
country_pop_gold['all_medals'] = country_pop_gold['gold'] + country_pop_gold['silver'] + country_pop_gold['bronze'] 
```

```python
rel = country_pop_gold.assign(medals_per_million = lambda df: df['all_medals']/ df.index.get_level_values(1))[['all_medals','medals_per_million']]
```

## *Final Table ranking by the top 25 countries

```python
rel.sort_values('medals_per_million', ascending=False).head(25)
```
